```python
import requests 
from bs4 import BeautifulSoup
import pandas as pd
import os
```

```python
# I scrape the website for information on specific tables for this project (ie. GDP, labour, corruption rank and index etc)
```

```python
# list of urls with the tables for the project 
urls = ["https://tradingeconomics.com/country-list/gdp?continent=africa",
       "https://tradingeconomics.com/country-list/external-debt?continent=africa",
       "https://tradingeconomics.com/country-list/personal-income-tax-rate?continent=africa", 
       "https://tradingeconomics.com/country-list/corruption-index?continent=africa",
       "https://tradingeconomics.com/country-list/government-debt-to-gdp?continent=africa"] 

```

```python
# I need to store the information from the parse process to excel files in my computer 


# directory to save the csv files 
save_directory ="D:\Africa Data"

# Ensure the directory exists
if not os.path.exists(save_directory):
    os.makedirs(save_directory)
```

```python
# create a dictionary to store all the tables by URL 
tables_dict = {}        
       

# due to webpage restrictions, custom headers were integrated to mimic browser request 
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}


# a loop function for each URL
for url in urls: 
    # GET request for the tables with headers 
    response = requests.get(url, headers=headers) 
    
    #error handling: to check if the request was successful 
    
    if response.status_code == 200:
        tables = pd.read_html(response.text)  
        tables_dict[url] = tables[0]

        #save the data into CSV File (excel) 
        table_name = url.split('?')[0].split('/')[-1]   # to use the url to name the tables
        file_path = os.path.join(save_directory, f"{table_name}_Africa.csv")

        
         # Save the tables to CSV in the specified directory
        tables_dict[url].to_csv(file_path, index=False)
        print(f"Table from {url} saved as '{file_path}'.")
    else: 
        print(f"Failed to retrieve the table. Status code: {response.status_code}")
print(tables_dict[urls[0]].head())
```

```python
# The first change in the table is standardizing the currency to USD and into actual values
```

```python
import pandas as pd


# Define conversion rates for currencies
conversion_rates = {
    "USD": 1,  # USD is the base
    "BWP": 0.071,  # Botswana Pula to USD
    "EUR": 1.03,  # Euro to USD
    "BIF": 0.00034,  # Burundian Franc to USD
    "KES": 0.0077, # Kenyan Shilling to USD
    "LSL": 0.05, # Lesotho Loti to USD
    "MLK": 0.00058, # Malawi Kwacha to USD 
    "MUR": 0.02136, # Mauritius Rupee to USD 
    "MAD": 0.09939, # Morocco Dirham to USD
    "NAD": 0.0560, # Namibia Dollar to USD 
    "SCR": 0.06958, # Seychellois Rupee to USD 
    "SZL": 0.05326, # Swaziland Lilangeni to USD
    "TND": 0.31096 # Tunisia Dollar to USD
    
}

# Function to calculate actual values
def calculate_actual_values(row):
    # Get the conversion rate for the currency
    conversion_rate = conversion_rates.get(row["Unit"], 1)  # Default to 1 if currency not found
    
    # Determine the multiplier based on Unit 2 Column
    if row["Unit 2"] == "Million":
        multiplier = 1_000_000
    elif row["Unit 2"] == "Billion":
        multiplier = 1_000_000_000
    else:
        multiplier = 1  # Default multiplier
    
    # Convert Last and Previous values
    last_value = row["Last"] * multiplier * conversion_rate
    previous_value = row["Previous"] * multiplier * conversion_rate
    
    return last_value, previous_value

# Apply the conversion to the DataFrame
df[["Last (USD)", "Previous (USD)"]] = df.apply(calculate_actual_values, axis=1, result_type="expand")

# Display the updated DataFrame
print(df)
```

```python
# Save the updated DataFrame to an Excel file
# directory to save the csv files 
save_directory ="D:\Africa Data"

# Ensure the directory exists
if not os.path.exists(save_directory):
    os.makedirs(save_directory)


# Save the updated DataFrame to a CSV file
output_path = "D:/Africa Data/updated_external_debt_africa_data.csv"  
df.to_csv(output_path, index=False)
```
